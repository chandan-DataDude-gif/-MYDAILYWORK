Handling the null value

```python
titanic_data=titanic_data.drop(columns ='Cabin',axis =1)
```

```python
titanic_data.info()
```

Replacing the null values of the 'Age'by mean of the age value

```python
titanic_data['Age']=titanic_data['Age'].fillna(titanic_data['Age'].mean())
```

Replacing the null value of the embarkment by the mode of the embarkment value

```python
titanic_data['Embarked'].mode()
```

```python
titanic_data['Embarked']=titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0])
```

```python
#replacing the value of the male and female by 0 and 1 and also replacing the Embarked value
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
```

```python
titanic_data.head(5)
```

```python
#Separating feature and target
X=titanic_data.drop(columns=['PassengerId','Survived','Name','Ticket'],axis=1)
Y=titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_test,X_train,Y_test,Y_train=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL EVALUATION

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#accuracy score on the train data
x_train_prediction=model.predict(X_train)
print(x_train_prediction)
```

ACCURACY SCORE

```python
from sklearn.metrics import accuracy_score
trainig_data_accuracy=accuracy_score(Y_train,x_train_prediction)
print('Accuracy score of the training data : ',trainig_data_accuracy)
```

```python
#accuracy score on the test data
y_test_prediction = model.predict(X_test)
print(y_test_prediction)
```

```python
test_data_accuracy  = accuracy_score(Y_test,y_test_prediction)
print('Accuracy_score of the data : ', test_data_accuracy)
```
